```python
sns.set()
```

```python
df_alex = pd.read_csv(SOURCE_DATA_ROOT / 'filter-similarities-alexnet.csv')
```

```python
for layer_name in df_alex.layer_name.unique():
    layer_df = df_alex[df_alex.layer_name == layer_name]
    g = sns.displot(
        data=layer_df, 
        x='similarity',
        hue='stim_type',
        col='layer_name',
        element='step',
        bins=50,
        stat="density", 
        common_norm=False,
        fill=False,
    )
```

```python
df_vgg16 = pd.read_csv(SOURCE_DATA_ROOT / 'filter-similarities-VGG16.csv')
```

```python
for block_num in df_vgg16.block_num.unique():
    block_df = df_vgg16[df_vgg16.block_num == block_num]

    g = sns.displot(
            data=block_df, 
            x='similarity',
            hue='stim_type',
            col='layer_num',
            row='net_number',
            element='step',
            bins=50,
            stat="density", 
            common_norm=False,
            fill=False,
    )
```
